### E20nM-F50nM-ROX

```python
import numpy as np
import matplotlib.pyplot as plt
from subtraction_catalysis import reaction

R = reaction()

C_N = 200
C_E = 20
C_F = 50
C_NEF = 0
C_W1 = 0
C_RE = 200
C_WE = 0
C_Re = 0
C_FuelE = 200
C_W2 = 0
C_RF = 200
C_WF = 0
C_Rf = 0
C_FuelF = 200
C_W3 = 0

C0 = [C_N, C_E, C_F, C_NEF, C_W1, C_RE, C_WE, C_Re, C_FuelE, C_W2, C_RF, C_WF, C_Rf, C_FuelF, C_W3,]

t = [0,120*60]
t_eval = np.linspace(t[0], t[1], 51)

R.NEF = 5e-6
R.NEFW = 5e-11
R.ERE = 3.9e-6 ##
R.FRF = 6.9e-6 ##
R.WERe = 5.55e-7
R.WFRf = 8.19e-7
R.WEF = 1.4e-5 ##
R.WFF = 2.9e-5 ##
R.EW = 5.97e-9
R.FW = 3.74e-7

R.T = 1
C_output = R.reaction(t, C0, t_eval)

R.T = 7
C_output1= R.reaction(t, C0, t_eval)

plt.plot(C_output.t, C_output.y[12]/200,'--',linewidth=2.0, color='red')
plt.plot(C_output1.t, C_output1.y[12]/200,'--',linewidth=2.0, color='blue')

plt.ylim([0,1])
plt.xlim([0,120*60])
plt.xlabel('Time (sec)')
plt.ylabel('Nor. Fluo.')
plt.legend(['Localized', 'Non_Localized'])
plt.title('E20nM-F50nM-ROX')
plt.show()
```

### Addition_A & B as an example

```python
import numpy as np
import matplotlib.pyplot as plt
from multiplication import reaction

A = reaction()

A.W = 2
A.I = 5e-4
A.M = 5e-4
A.aW = 3e-11
A.S = 5e-4
A.fS = 3e-11

C_I = 20
C_H = 25
C_O = 5
C_W = 0
C_W1 = 0
C_M = 25*A.W
C_aW = 0
C_W2 = 0
C_S = 150
C_fS = 0
C_E = 0

t = [0,120*60]
t_eval = np.linspace(t[0], t[1], 51)

# A

C0 = [C_I, C_H, C_O, C_W, C_W1, C_M, C_aW, C_W2, C_S, C_fS, C_E,]
A.T = 1
C_output = A.reaction(t, C0, t_eval)
A.T = 2
C_output1= A.reaction(t, C0, t_eval)

# B

A.W = 4
C_M = 25*A.W

C1 = [C_I, C_H, C_O, C_W, C_W1, C_M, C_aW, C_W2, C_S, C_fS, C_E,]
A.T = 1
C_output2 = A.reaction(t, C1, t_eval)
A.T = 2
C_output3= A.reaction(t, C1, t_eval)

plt.plot(C_output.t, (C_output.y[10]+C_output2.y[10])/200,'--',linewidth=2.0, color='red')
plt.plot(C_output1.t, (C_output1.y[10]+C_output3.y[10])/200,'--',linewidth=2.0, color='blue')

plt.ylim([0,1])
plt.xlim([0,120*60])
plt.xlabel('Time (sec)')
plt.ylabel('Nor. Fluo.')
plt.legend(['Localized', 'Non_Localized'])
plt.title('addition')
plt.show()
```
